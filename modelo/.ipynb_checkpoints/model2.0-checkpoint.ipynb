```python
import pandas as pd

with open("Comentarios.csv", "r", encoding="latin-1") as file:
    # Lee todas las líneas, eliminando espacios en blanco, y descarta las vacías.
    comments = [line.strip() for line in file if line.strip()]

# Limpiar comillas dobles "" de cada elemento de la lista
for i in range(len(comments)):
    # Si en la línea hay comillas dobles, las reemplazamos por vacío
    if '"' in comments[i]:
        comments[i] = comments[i].replace('""', "")
    # Limpiamos espacios en blanco sobrantes
    comments[i] = comments[i].strip()

# Creamos listas para comentarios y sentimientos
lista_comentarios = []
lista_sentimientos = []

# Separamos cada línea en comentario y sentimiento
for linea in comments:
    partes = linea.split(":")
    
    # Verifica que la línea tenga al menos dos partes
    if len(partes) < 2:
        print("Línea con formato inesperado:", linea)
        continue

    comentario = partes[0].strip()
    sentimiento = partes[1].strip()
    print(sentimiento)
    lista_comentarios.append(comentario)
    lista_sentimientos.append(sentimiento)


df = pd.DataFrame()
df["comentario"] = lista_comentarios
df["sentimiento"] = lista_sentimientos
df.sentimiento = df.sentimiento.apply(lambda x : x.replace('"', "").strip())
df.comentario = df.comentario.apply(lambda x : x.replace('"', "").strip())
df.head(100
)
```

```python
def es_fila_correcta(row):
    return row["sentimiento"] in ["halago", "info", "tecnico"]

df = df[df.apply(es_fila_correcta, axis=1)]
```

```python
#Dividir el conjunto de datos(train, validation, test)
from sklearn.model_selection import train_test_split

train_validation, test_df = train_test_split(df, test_size = 0.10)
train_df, validation_df = train_test_split(train_validation, test_size = 0.15/(1-0.10))

train_df = train_df.reset_index()
validation_df = validation_df.reset_index()
test_df = test_df.reset_index()
```

```python
print("TOTAL TRAIN:",train_df.shape[0])
print("TOTAL VALIDATION:",validation_df.shape[0])
print("TOTAL TEST:",test_df.shape[0])
```

```python
train_df.drop(columns = "index", inplace = True)
validation_df.drop(columns = "index", inplace = True)
test_df.drop(columns = "index", inplace = True)
```

```python
train_df.head()
```

```python
#Codificación One-Hot de los conjuntos de entrenamiento/validación/prueba
train_df = pd.get_dummies(train_df, columns=["sentimiento"])
validation_df = pd.get_dummies(validation_df, columns=["sentimiento"])
test_df = pd.get_dummies(test_df, columns=["sentimiento"])
```

```python
y_train = train_df[["sentimiento_halago", 
                    "sentimiento_info", 
                    "sentimiento_tecnico"]].values
y_val = validation_df[["sentimiento_halago", 
                    "sentimiento_info", 
                    "sentimiento_tecnico"]].values
y_test = test_df[["sentimiento_halago", 
                    "sentimiento_info", 
                    "sentimiento_tecnico"]].values
```

```python
# Suponiendo que ya dividiste tu dataset en train_df, val_df, test_df
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 1. Crear y ajustar el Tokenizer con los datos de entrenamiento
tokenizer = Tokenizer(num_words=5000, oov_token='<OOV>')
tokenizer.fit_on_texts(train_df["comentario"])

# 2. Convertir a secuencias en cada subset
train_sequences = tokenizer.texts_to_sequences(train_df["comentario"])
val_sequences = tokenizer.texts_to_sequences(validation_df["comentario"])
test_sequences = tokenizer.texts_to_sequences(test_df["comentario"])

# 3. Aplicar padding
max_length = 50
train_padded = pad_sequences(train_sequences, maxlen=max_length, padding='post', truncating='post')
val_padded = pad_sequences(val_sequences, maxlen=max_length, padding='post', truncating='post')
test_padded = pad_sequences(test_sequences, maxlen=max_length, padding='post', truncating='post')
```

```python
X_train = train_padded
X_val = val_padded
X_test = test_padded
```

```python
train_padded
```

```python
vocab_size = 5000          # Debe ser el mismo que usaste en el Tokenizer
embedding_dim = 128        # Dimensión del vector de embedding (puedes ajustar)
max_length = 50            # La misma longitud que usaste para el padding
```

```python
import pickle

with open("modeltokenizer.pickle", "wb") as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense

model = Sequential()
# Capa de Embedding
model.add(Embedding(input_dim=vocab_size,
                    output_dim=embedding_dim,
                    input_length=max_length))

# Primera capa LSTM bidireccional (32 unidades en lugar de 64)
model.add(Bidirectional(LSTM(32, return_sequences=True)))

# Segunda capa LSTM bidireccional (32 unidades nuevamente)
# return_sequences=False para que devuelva solo el último estado
model.add(Bidirectional(LSTM(32, return_sequences=False)))

# Dropout para prevenir sobreajuste
model.add(Dropout(0.5))

# Capa de salida
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=15,
    batch_size=32,
    validation_data=(X_val, y_val)
)
```

```python
import matplotlib.pyplot as plt

# Obtener los datos del historial
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# Graficar la precisión en entrenamiento y validación
plt.figure(figsize=(32, 16))  # Aumentar el tamaño de la figura

plt.subplot(1, 2, 1)
plt.plot(epochs, acc, 'bo-', label='Entrenamiento', markersize=10)  # Aumentar el tamaño de los marcadores
plt.plot(epochs, val_acc, 'ro-', label='Validación', markersize=10)  # Aumentar el tamaño de los marcadores
plt.title('Precisión durante el entrenamiento y la validación', fontsize=24)  # Aumentar el tamaño del título
plt.xlabel('Épocas', fontsize=20)  # Aumentar el tamaño de la etiqueta del eje x
plt.ylabel('Precisión', fontsize=20)  # Aumentar el tamaño de la etiqueta del eje y
plt.legend(fontsize=18)  # Aumentar el tamaño de la leyenda
plt.xticks(fontsize=16)  # Aumentar el tamaño de las etiquetas de los ticks en el eje x
plt.yticks(fontsize=16)  # Aumentar el tamaño de las etiquetas de los ticks en el eje y

# Graficar la pérdida en entrenamiento y validación
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, 'bo-', label='Entrenamiento', markersize=10)  # Aumentar el tamaño de los marcadores
plt.plot(epochs, val_loss, 'ro-', label='Validación', markersize=10)  # Aumentar el tamaño de los marcadores
plt.title('Pérdida durante el entrenamiento y la validación', fontsize=24)  # Aumentar el tamaño del título
plt.xlabel('Épocas', fontsize=20)  # Aumentar el tamaño de la etiqueta del eje x
plt.ylabel('Pérdida', fontsize=20)  # Aumentar el tamaño de la etiqueta del eje y
plt.legend(fontsize=18)  # Aumentar el tamaño de la leyenda
plt.xticks(fontsize=16)  # Aumentar el tamaño de las etiquetas de los ticks en el eje x
plt.yticks(fontsize=16)  # Aumentar el tamaño de las etiquetas de los ticks en el eje y

plt.tight_layout()  # Asegurar que no haya superposición entre las subgráficas
plt.show()
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=2)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
